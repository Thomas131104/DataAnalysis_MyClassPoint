```python
import pandas as pd
import pdfplumber

# Mở file của thầy
with pdfplumber.open("Diem thi QLLV.pdf") as pdf:
    page1 = pdf.pages[0]
    table1 = page1.extract_table()
    HM = pd.DataFrame(table1[1:], columns = table1[0])
    #  Đổi tên cột
    HM = HM.rename(
        columns = {
            'Chuyên cần LT (10-\nsố vắng mặt LT*2)' : 'CC_LT',
            'Chuyên cần TH\n(số bài TH*2' : 'CC_TH',
            'Chuyên cần (20%) =\n(Chuyên cần LT + TH)/2' : 'CC',
            'Thi GK (20%)' : 'GK',
            'Thi CK (60%)' : 'CK',
            'Điểm TK = (Chuyên cần +\nThi GK)*0.2+Thi CK*0.6' : 'TK10'
        }
    )

    # Loại bỏ các cột không cần thiết
    HM = HM.drop(columns = ['Stt', 'Họ lót', 'Tên', 'Email'])
    HM.to_excel("Data\\DH22HM.xlsx", index = False)


    # Làm tương tự với lớp DH22DL
    page2 = pdf.pages[1]
    table2 = page2.extract_table()
    DL = pd.DataFrame(table2[1:], columns = table2[0])
    #  Đổi tên cột
    DL = DL.rename(
        columns = {
            'Vắng mặt L' : 'Vắng mặt LT',
            'Chuyên cần LT (10-\nTsố vắng mặt LT*2)' : 'CC_LT',
            'Chuyên cần TH\n(số bài TH*2' : 'CC_TH',
            'Chuyên cần (20%) =\n(Chuyên cần LT + TH)/2' : 'CC',
            'Thi GK (20%)' : 'GK',
            'Thi CK (60%)' : 'CK',
            'Điểm TK = (Chuyên cần +\nThi GK)*0.2+Thi CK*0.6' : 'TK10'
        }
    )

    # Loại bỏ các cột không cần thiết
    DL = DL.drop(columns = ['Stt', 'Họ lót', 'Tên', 'Email'])
    DL.to_excel("Data\\DH22DL.xlsx", index = False)    
```
