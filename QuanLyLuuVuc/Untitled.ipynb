```python
class Point:
    def __init__(self, x, y):
        self._x = x
        self._y = y

    @property
    def x(self): return self._x

    @property
    def y(self): return self._y

    @x.setter
    def x(self, value): 
        self._x = value

    @y.setter
    def y(self, value): 
        self._y = value

    def __str__(self): return f"Point({self.x}, {self.y})"
```

```python
import random

POINT = []
N = 1000
for _ in range(N):
    x = random.uniform(-100, 100)
    y = random.uniform(-100, 100)
    POINT.append(Point(x, y))

for i in range(5):
    print(POINT[i])
```

```python
POINT_X = [point.x for point in POINT]
POINT_Y = [point.y for point in POINT]
```

```python
import matplotlib.pyplot as plt

# To kch thc ca  th
plt.figure(figsize = (10, 10))

# To biu  im
plt.scatter(POINT_X, POINT_Y, s=10, c='blue', alpha=0.5)

# My ci trang tr
plt.xlabel("Trc X")
plt.ylabel("Trc Y")
plt.title("S lng im")
plt.axis([-100, 100, -100, 100])
plt.show()
```
